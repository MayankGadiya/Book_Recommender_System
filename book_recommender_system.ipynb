## Popularity based recommender system

```python
# merging two df books and ratings
ratings_with_name = ratings.merge(books, on= 'ISBN')
```

```python
ratings_with_name.head(5)
```

```python
# creating df based on book title and no of ratings
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace = True)
num_rating_df
```

```python
# craeting df on book title and avg ratings
# converting book ratings column to float as some values are object type
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace = True)
avg_rating_df
```

```python
# merging both dataframes i.e num of ratings df and avg ratings df
popular_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popular_df
```

```python
# keeping only those books whose no of ratings are more than or equal to 250
popular_df[popular_df['num_ratings']>=250]
```

```python
# sort based on avg ratings
popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending  = False)
```

```python
# top 50 books
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending  = False).head(50)
```

```python
popular_df.head()
```

```python
#merging this popular df with books to get author name, image url 
# as there many entries for single book in books df we will drop duplicates based on book title and get final top 50 books
popular_df = popular_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_ratings']]
```

# Collaborative Filtering Based Recommendation System

```python
#now for this type of sysem we will find those users who have rated for more then 200 books , so groupby using user-id and finding count 
x= ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users= x[x].index
```

```python
#now keeping inly those userid which we got in x
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating.head()
```

```python
#now we will keep only those books which are been rated atleast by 50 users
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>= 50
```

```python
famous_books= y[y].index
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)].head()
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# creating pivot table
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
#filling na values with 0
pt.fillna(0, inplace = True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True) [1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```
